#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
def normalizar(group):
    mean_price = group["price"].mean()
    std_dev_price = group["price"].std()
    precio_normal = (group["price"]-mean_price) / std_dev_price
    return precio_normal
normal_prices = south_belt.groupby("level5").apply(normalizar).rename("precios normales").reset_index()
print(normal_prices)
for name, group in normal_prices.groupby("level5"):
    plt.hist(group['precios normales'], bins=30, alpha=0.5, label=f'Grupo {name}')

plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.title('Histogramas de Precios Normalizados por Grupo')
plt.legend()
plt.grid(True)
plt.show()

```

```python
from ipyleaflet import Map, Marker, AwesomeIcon, basemaps

# Mapa centrado en Madrid
center = [40.35, -3.75]
zoom = 10
map = Map(basemap=basemaps.CartoDB.Positron, center=center, zoom=zoom)

# Diccionario de colores
colors = {
    'Getafe': "red",
    'Leganés': "blue",
    'Fuenlabrada': "green",
    'Alcorcón': "purple",
}

# Iterar sobre cada fila del DataFrame
for index, row in south_belt.iterrows():
    lat = row["latitude"]
    lon = row["longitude"]
    population = row["level5"]

    # Obtener el color correspondiente o usar negro por defecto
    color = colors.get(population, "black")

    # Crear un marcador con el ícono y color correspondiente
    icon = AwesomeIcon(name='info-circle', marker_color=color, icon_color="white")
    marker = Marker(location=(lat, lon), icon=icon)
    
    # Agregar el marcador al mapa
    map.add_layer(marker)

# Mostrar el mapa
map
```
